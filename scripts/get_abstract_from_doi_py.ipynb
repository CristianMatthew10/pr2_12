```python
# Clean abstract, removing html tags
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
    cleantext = re.sub(CLEANR, ' ', raw_html)
    return cleantext

def abstract(x):
    html_text = works.doi(x)
    for item in html_text.keys():
        if item == 'abstract':
            return cleanhtml(html_text['abstract'])
    return 'Empty'

refs['abstract']=refs['URLs'].apply(abstract)
```
